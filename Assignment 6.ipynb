```python
#2(a). Heat capacity of a solid
import numpy as np
def heat_capacity(T):
    rho = 6.022*10**28
    theta_d = 428 
    V = .001
    K_b = 1.38064852*10**(-23)
    N = 1000
    a = 0.0
    b = (theta_d / T)
    h = (b-a)/N
    s= 0.5*(((b**4)*(np.exp(b)))/((((np.exp(b))-1)**2)))
    for k in range (1, N):
        s += ((((a+k*h)**4)*(np.exp(a+k*h)))/(((np.exp(a+k*h))-1)**2))
    Cv = 9*V*rho*K_b*((T/theta_d)**3)*(s*h)
    return (Cv)
```

```python
#check that the function works:
heat_capacity(5)
```

```python
#putting these values into a list
temp = []
for x in range(4,500):
    temp.append(x)
temp
```

```python
HC = []
for x in range(4,500):
    HC.append(heat_capacity(x))
HC
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(temp, HC, c='y')
ax.set_xlabel("Temperature")
ax.set_ylabel("Heat capacity")
```
